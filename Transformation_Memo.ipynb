# Transformation Memo: Board of Directors, Kaspi.kz

**To:** Board of Directors, Kaspi.kz  
**From:** Product Operating Model Consultant  
**Date:** 2026-01-24  
**Subject:** Urgent Pivot Required: Escaping the Feature Factory Trap

## 1. Executive Summary
Kaspi.kz is currently operating as a high-functioning **Feature Factory**, achieving scale through sheer execution force rather than sustainable product innovation. While financial results are strong, the organizational reliance on "IT" service delivery and business-appointed leadership creates a fragile structure of mercenaries that will fail to innovate as market complexity increases.

## 2. Evidence of Feature Factory Behavior
Despite external success, internal signals reveal an output-centric culture inconsistent with the Product Operating Model:

*   **Engineering as a Service (IT Mindset):**
    > "One of our main recruitment priorities is entry level IT positions." (ESG Report, p.31)
    *Signal:* Referring to engineers as "IT" frames them as back-office cost centers (Mercenaries) rather than the core business value creators. In a product company, technology *is* the business.

*   **Stakeholder-Driven Roadmap (Lack of Empowerment):**
    > "Each team has a leader, or product head, who is usually appointed from within the business... This approach helps us make decisions quickly." (ESG Report, p.30)
    *Signal:* Appointing leaders "from the business" typically delegates engineers to order-takers. True empowerment requires skilled Product Managers who discover *value*, not just Project Managers who deliver *requests*.

*   **Output-Centric Competence:**
    > "Kaspi Lab... master applied skills in 7 areas... Oracle, MS SQL, Testing and accounting." (ESG Report, p.32)
    *Signal:* Training focuses entirely on *delivery tools* (SQL, Oracle) and *compliance* (Accounting). There is zero mention of Product Discovery, User Research, or Prototyping—skills required to figure out *what* to build, not just *how* to build it.

## 3. Strategic Disconnects
*   **The "Super App" Trap:** The strategy relies on bundling services ("synergetic with everything else") rather than ensuring each product solves a distinct, viable customer problem on its own. This masks poor product performance with forced ecosystem integration.
*   **Mercenary Culture disguised as Mission:** While the "mission" is stated, the structural reality (IT Dept + Business Owners) creates a culture where success is measured by "shipping features" (e.g., Digital Documents, Tax Reports) rather than solving hard problems.
*   **GovTech as Vanity:** High usage of government services ("11 million users") is cited as success, but this is often a captive audience necessity, not a validation of product-market fit or user love.

## 4. The Prescription
To transition from a Feature Factory to an Empowered Product Organization, the Board must mandate:

*   **Transform "IT" into Product Engineering:** Abolish the concept of "IT". Reorganize into durable, cross-functional Product Teams where engineers sit with designers and are measured by *business outcomes*, not "projects delivered".
*   **Install Competent Product Management:** Stop appointing "business heads" as proxy product owners. Hire and train true **Product Managers** capable of running Product Discovery—assessing Value, Usability, Feasibility, and Viability *before* code is written.
*   **Fund Outcomes, Not Projects:** Shift the budgeting model from funding specific deliverables (e.g., "Build the Tax Report feature") to funding business objectives (e.g., "Reduce Small Business Tax Compliance Friction by 50%"). Give teams the autonomy to find the best solution.
