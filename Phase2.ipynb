## **Phase 2 — JTBD Discovery**

### Role: Product Manager / JTBD Lead
**Company: Kaspi.kz**

**Domain: Procurement & Legal Operations**

### **2.1 Purpose of Phase 2**

The purpose of Phase 2 is to move beyond high-level organizational diagnosis and identify a specific, actionable internal problem that can be solved with an AI-enabled product.

Using the Jobs to be Done (JTBD) framework, this phase simulates a real internal user and applies the 5 Whys technique to uncover the root cause behind surface-level inefficiencies.
The goal is not to justify AI usage, but to determine whether AI meaningfully reduces risk, fear, or friction in a critical internal workflow.

________________________________________

### **2.2 Identified Problem (Derived from Phase 1 Insights)**

Based on the Strategic Diagnosis (Phase 1), Kaspi.kz demonstrates high digital maturity at the product level.
However, **internal control-heavy functions (Procurement / Legal)** remain constrained by:

- High audit pressure

- Strict compliance accountability

- Fear of personal responsibility for contractual errors

This suggests a gap between **organizational digital maturity** and **internal operational confidence**.

### **2.3 Simulated User Persona (AI Persona)**

#### *AI Persona Profile*

Role: Senior Procurement Officer  
Company: Kaspi.kz  
Experience: 8+ years in procurement and contract review  
Context:
- Works with supplier and service contracts daily
- Operates under constant audit and compliance oversight
- Final approval decisions carry personal accountability

**Primary Pain:**
Fear of approving incorrect contract clauses that may trigger audits,
penalties, or reputational damage.

**Motivation:**
Avoid mistakes and personal liability, even at the cost of speed.

**Current Behavior:**
Prefers manual, repetitive review processes over automation due to
lack of trust and decision support.

### **2.4 JTBD Interview Log (5 Whys)**

| Step | Product Manager Question | AI Persona Response |
|----|--------------------------|---------------------|
| 1 | Why does the procurement approval process take so long? | Because every contract must be manually reviewed in detail before approval. |
| 2 | Why is manual review required for every contract? | Because even small mistakes can lead to audits, penalties, or internal investigations. |
| 3 | Why is there no reliance on templates or partial automation? | Templates do not explain risks, and automation does not protect me from responsibility if something goes wrong. |
| 4 | Why is personal responsibility such a strong blocker? | Accountability is individual. If an error is found, the approving officer is responsible, not the system. |
| 5 | Why is fear stronger than efficiency in this process? | There is no safety net — no decision support that explains risks, flags issues clearly, or supports my judgment. |

### **2.5 Key Insight (Root Cause)**

> The core problem is not speed or workload.
> 
> It is fear of personal accountability caused by the absence of explainable decision support.

Procurement officers do not resist automation because they are inefficient,
but because **automation does not reduce personal risk.**
This shifts the problem framing from operational efficiency to psychological
and governance-driven risk avoidance.

________________________________________

### **2.6 Job to Be Done (JTBD Statement)**

When reviewing procurement and legal contracts,
I want clear, explainable risk highlighting and policy references,
so that I can approve decisions faster without fear of violating compliance
or exposing myself to personal liability.

### **2.7 Product Requirements Document (PRD)**

**Problem Statement**

Procurement and legal professionals at Kaspi.kz spend excessive time manually reviewing contracts due to fear of compliance violations and personal accountability. Existing tools do not provide sufficient risk transparency or decision support.



**Target User**

 - Procurement Officers

 - Legal Review Specialists

 - Compliance Teams (read-only)



**Functional Requirements**

1. Clause-Level Risk Highlighting
   - Identify potentially risky clauses
   - Highlight deviations from internal standards

2. Policy and Regulation Referencing
   - Link flagged clauses to internal policies or rules
   - Provide source-based explanations

3. Explainability
   - Clearly explain WHY a clause is flagged
   - No black-box decisions

4. Read-Only Decision Support
   - AI does not approve or reject contracts
   - AI provides guidance only



**Non-Goals**

- Autonomous contract approval
- Legal decision-making without human review
- Modification of contract text



**Constraints & Governance**

- Human-in-the-loop mandatory
- Read-only access to documents
- No long-term document storage
- Compliance with Kazakhstan data residency requirements
- Full logging and auditability of AI outputs

### **2.8 Phase 2 Output Summary**

This phase demonstrates that the proposed AI Agent is not designed to replace human judgment, but to **reduce cognitive load** and **compliance anxiety** by providing transparent, explainable decision support.

The findings of Phase 2 directly inform:

 - The **ROI calculation** in Phase 3 (time savings)

 - The **Governance controls** required for safe deployment
