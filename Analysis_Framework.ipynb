# Product Operating Model Audit: Kaspi.kz

**Date:** 2026-01-24
**Subject:** Organizational Maturity Audit (Marty Cagan’s Model)
**Source:** Kaspi.kz 2023 ESG Report (PR_20-F_2023.txt was insubstantial)

## Executive Summary
Kaspi.kz exhibits a high degree of Product Operating Model maturity. The organization is structured around empowered product teams ("Missionaries") with a clear focus on business and social outcomes rather than mere output. Technology is positioned as the core business enabler, not a support function, though some traditional nomenclature ("IT") persists in recruitment and events.

---

## 1. Mercenaries vs Missionaries
**Assessment: Missionaries (High Maturity)**
Kaspi.kz empowers teams with clear ownership and a strong mission connection.

*   **Evidence:** "Kaspi.kz’s mission is to improve people’s lives by developing innovative mobile products and services." (Page 3)
*   **Evidence:** "Each team has a leader, or product head, who is usually appointed from within the business and is responsible for a specific product. This approach helps us make decisions quickly... For employees the benefits are increased responsibility, clear expectations and accelerated career growth." (Page 30)
    *   *Analysis:* This explicitly describes the "Empowered Product Team" model where a leader owns the product and decisions, fostering specific accountability (Missionaries) rather than taking orders (Mercenaries).
*   **Evidence:** "Direct responsibility is fundamental to the way we work... The concept of directly responsible individuals... refers to the person, with whom accountability for any given product sits." (Page 30)

## 2. Project vs Product Funding
**Assessment: Product Funding (Medium-High Maturity)**
While the finance mechanics are not detailed, the strategy reflects continuous investment in value streams and "capex lite" scaling rather than discrete, heavy project budgeting.

*   **Evidence:** "We operate a unique, two-sided Super App model... best-in-class, highly relevant digital products and a capex lite approach." (Page 3)
    *   *Analysis:* "Capex lite" implies avoiding the "Project" trap of heavy upfront capitalization, favoring continuous digital iteration (Opex/Product funding).
*   **Evidence:** "We immediately make our new GovTech products available via open platform... Although we do not generate revenue directly from GovTech services, they’re synergetic with everything else." (Page 18)
    *   *Analysis:* Investing in non-revenue generating products for ecosystem synergy is a hallmark of strategic Product Funding, whereas Project Funding would likely kill such initiatives for lack of direct ROI.
*   **Evidence:** "We have once again allocated substantial funds for multiple charitable initiatives." (Page 19)
    *   *Analysis:* Shows ability to allocate funds to strategic themes (Outcomes) rather than just revenue features.

## 3. IT vs Product
**Assessment: Integrated Product Company (High Maturity)**
Technology IS the business at Kaspi.kz. There is no evidence of "IT" being a subservient order-taker for "The Business".

*   **Evidence:** "One of our main recruitment priorities is entry level IT positions." (Page 31)
    *   *Analysis:* Uses "IT" as a functional discipline label for hiring, but placing them in product teams (see Section 1) negates the "IT Dept" service-bureau risk.
*   **Evidence:** "Our proprietary data... gives us unique and superior insights... continuously improve with each additional transaction... We can add a new data point within one second and fully update our models in less than one hour." (Page 21)
    *   *Analysis:* This speed and integration demonstrate that engineering/data is the core competency of the company.
*   **Evidence:** "We work hand in hand with various government agencies... GovTech is another channel for us to digitalise everyday life." (Page 17)
    *   *Analysis:* Tech strategy is indistinct from Business strategy.

## 4. Output vs Outcome
**Assessment: Outcome Focused (High Maturity)**
The organization relentlessly measures value delivered to customers and the ecosystem, not just features shipped.

*   **Evidence:** "The ratio of average daily to average monthly users is 65%, which is amongst the highest of any Super App globally." (Page 9)
    *   *Analysis:* Focus on *Engagement* (Outcome) rather than just number of users or downloads (Vanity interactions).
*   **Evidence:** "In 2023 we delivered 43 million orders... with 88% of orders delivered free for the consumer." (Page 4)
    *   *Analysis:* Measures the *value* (free delivery, successful orders) to the user.
*   **Evidence:** "We design and regularly assess our information security program... In 2023 no single cybersecurity incidents were registered." (Page 22)
    *   *Analysis:* Security is measured by the *Outcome* (safe platform) not just compliance outputs.
*   **Evidence:** "Transactions cancelled due to the suspicion of fraud <0.1%." (Page 23)
*   **Evidence:** "Net Promoter Score (NPS) framework... average eNPS score for our senior team was 83%." (Page 34)
