Phase 3 — Managerial Defense (Go / No-Go Decision)
Role: Strategy, Finance & Governance Lead
________________________________________
1. Purpose of This Phase
The purpose of Phase 3 is to make a clear, accountable managerial decision regarding the proposed Internal AI Agent for Procurement and Legal operations at Kaspi.kz.
While Phase 1 established the organizational and strategic context, and Phase 2 identified a real internal problem using the Jobs-to-be-Done (JTBD) framework, this phase addresses the final executive question: should this solution actually be built?
This phase evaluates the proposal using two decision-making criteria that are critical for senior management:
1.	Whether the solution is financially justified and sustainable.
2.	Whether the solution can be implemented in a safe, ethical, and legally compliant manner within the Kazakhstan regulatory environment.
________________________________________
2. Financial Reasoning (ROI Analysis)
2.1 Assumptions (Kazakhstan Context)
The financial analysis is based on conservative and realistic assumptions reflecting the Kazakhstan labor market and Kaspi.kz’s internal operational roles.
The primary users of the AI Agent are assumed to be procurement specialists, legal officers, and compliance professionals who regularly work with contracts, regulatory documents, and internal approval processes. A significant portion of their work involves manual document review, repetitive checks, and risk scanning.
Assumptions:
•	Average monthly salary: 450,000 KZT
•	Standard workload: 160 hours per month
•	Approximate hourly cost: ≈ 2,800 KZT/hour
These assumptions are intentionally conservative and suitable for a managerial-level evaluation rather than optimistic forecasting.
________________________________________
2.2 Estimated Productivity Gains
Based on the PRD defined in Phase 2, the AI Agent is expected to save up to 10 working hours per week per employee by accelerating document review, highlighting potential risks, and reducing repetitive manual checks.
Calculation:
•	Time saved per year: 10 hours/week × 52 weeks = 520 hours
•	Annual productivity gain per employee:
520 hours × 2,800 KZT ≈ 1,456,000 KZT
This productivity gain does not imply staff reduction, but rather a reallocation of time toward higher-value activities such as decision-making, risk evaluation, and cross-functional coordination.
________________________________________
2.3 Estimated AI Operating Cost
The estimated annual operating cost of the AI Agent for a pilot deployment includes:
•	LLM usage (token-based inference)
•	Basic infrastructure and hosting
•	Monitoring, logging, and access control for governance
For a limited pilot implementation, the total annual operating cost is estimated at 300,000–500,000 KZT.
This controlled cost structure allows Kaspi.kz to validate value before any scaling decision, minimizing financial risk.
________________________________________
2.4 ROI Snapshot (Pilot Scenario)
Pilot scale assumption: 10 employees
•	Annual benefit:
1,456,000 KZT × 10 employees ≈ 14,560,000 KZT
•	Annual operating cost (upper estimate): 500,000 KZT
•	Net annual benefit: ≈ 14,060,000 KZT
Sensitivity check:
Even if actual time savings are only 5 hours per week, the annual benefit remains approximately 728,000 KZT per employee, which still significantly exceeds operating costs.
Financial conclusion:
The investment pays back well within one year and represents a low-risk, high-return pilot initiative.
________________________________________
3. AI Governance & Ethics (Kazakhstan Context)
Kaspi.kz operates in a highly regulated financial environment and already demonstrates strong practices in governance, cybersecurity, and data protection. However, introducing an internal AI agent introduces new categories of risk that must be explicitly managed.
________________________________________
3.1 Data Residency and Privacy
Procurement and legal documents may contain personal and confidential information. Kazakhstan regulations require careful handling and localization of personal data.
Risk mitigation measures:
•	No permanent document storage by the AI Agent
•	Session-based processing only
•	Masking of personal data where possible
•	Deployment aligned with Kazakhstan data residency requirements (on-prem or private cloud within Kazakhstan)
________________________________________
3.2 Corporate and Banking Secrecy
Unauthorized access to procurement or legal materials could lead to legal and reputational risks.
Controls:
•	Strict role-based access control (RBAC)
•	Read-only access for the AI Agent
•	AI acts strictly as a support tool
•	All approvals and final decisions remain with human employees
________________________________________
3.3 Hallucination Risk
AI systems may generate incorrect or misleading outputs when working with complex legal texts.
Mitigation approach:
•	AI positioned as a copilot, not a decision-maker
•	Mandatory human review of all outputs
•	Responses grounded only in provided source documents
•	No autonomous legal or compliance decisions
________________________________________
3.4 Accountability and Transparency
Accountability must always remain with people.
Governance measures:
•	Full logging and auditability of AI interactions
•	Clear ownership of decisions at the human role level
•	Transparent usage aligned with Kaspi.kz’s existing governance framework
________________________________________
4. Executive Decision
Final Recommendation: GO (Pilot Phase)
Based on the financial and governance analysis, the recommended managerial decision is to approve the project for a controlled pilot implementation.
Rationale:
1.	The ROI is clearly positive under conservative assumptions.
2.	Identified risks are well understood and manageable.
3.	The solution aligns with Kaspi.kz’s digital maturity and governance culture.
4.	There is no apparent conflict with Kazakhstan legal or ethical requirements, subject to standard legal and compliance review.
________________________________________
Conditions for Approval
•	Pilot limited to one department
•	Read-only AI access
•	Mandatory human approval for all outputs
•	Pre-launch legal and compliance review
•	Post-pilot evaluation before any scaling decision
________________________________________
5. Final Managerial Assessment
The proposed Internal AI Agent represents a financially justified, operationally controlled, and ethically governed initiative. The expected benefits significantly exceed the costs, while risks can be effectively managed through existing governance mechanisms.
Given Kaspi.kz’s experience in regulated operations, cybersecurity, and data governance, this initiative is suitable for pilot deployment and supports long-term improvements in internal efficiency without compromising compliance or accountability.
